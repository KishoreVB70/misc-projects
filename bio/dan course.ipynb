```python
import numpy as np
import matplotlib.pyplot as plt

# Data
one_over_v = [0.001716, 0.003486, 0.002299, 0.002445, 0.002053, 0.001717, 0.001236, 0.003286]
one_over_S = [100, 50, 33.33, 25, 20, 16.67, 14.29, 12.5]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(one_over_S, one_over_v, color='blue', label='Data Points')

# Linear regression
m, b = np.polyfit(one_over_S, one_over_v, 1)
plt.plot(one_over_S, m*np.array(one_over_S) + b, color='red', label='Linear Regression')

# Add labels and title
plt.xlabel('1/[S] (1/g/L)')
plt.ylabel('1/v (1/mol/(mLÂ·min))')
plt.title('Lineweaver-Burk Plot')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
